# 1. Limpieza de Datos

```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv('data/titanic_clean.csv')
```

```python
df.head()
```

```python
print("Valores nulos por columna")
print(df.isnull().sum())

print(f"El shape del conjunto de entrenamiento es {df.shape}. {df.shape[0]} filas y {df.shape[1]} columnas.") 

print("-------------------------------------------------")
df.info()
```

# Mapa de Calor de Valores Nulos

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Mapa de calor de los valores nulos del dataframe.
# df.isnull() devuelve True (1) donde hay nulos y False (0) donde hay datos,
# por lo que cada celda resaltada en el mapa indica un valor faltante.
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Mapa de calor de valores nulos')
plt.xlabel('Columnas')
plt.ylabel('Filas')
plt.show()
```

# Tratamiento de Nulos

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])


df.to_csv('data/titanic_clean.csv', index=False)
print("Datos guardados en data/titanic_clean.csv")
```
